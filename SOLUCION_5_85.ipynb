```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/data_wage_train.csv')
```

```python
test = pd.read_csv('/content/data_test.csv')
```

```python
y = df['wage']
x = df.drop(columns='wage')
```

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
import pandas as pd

param_grid = {'n_estimators': [100, 150, 200],
              'max_depth': [5, 8, 10],
              'learning_rate': [0.01, 0.1, 0.2]}

lgb_model = lgb.LGBMRegressor(random_state=42)

lgb_grid = GridSearchCV(estimator=lgb_model, param_grid=param_grid,
                        scoring='neg_mean_squared_error', cv=5)

lgb_grid.fit(x, y)

best_lgb_model = lgb_grid.best_estimator_

preds_test_lgb = best_lgb_model.predict(test)

submission_df_lgb = pd.DataFrame({'id': range(0, len(preds_test_lgb)),
                                   'Expected': preds_test_lgb})

submission_df_lgb.to_csv('sol_lgb.csv', index=False)
```
